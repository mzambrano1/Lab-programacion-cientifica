````python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
````

````python
from google.colab import drive6
# ejecutar en el caso de usar colab.
drive.mount("/content/drive")
````

````python
df = pd.read_csv("./water_potability.csv").dropna()
X = df.drop(columns=["Potability"])
y = df.loc[:, ["Potability"]].values.ravel()
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, shuffle=True, stratify=y
)

ros = RandomOverSampler(random_state=42)
rus = RandomUnderSampler(random_state=42)
X_train, y_train = ros.fit_resample(X_train, y_train)
X_train, y_train = rus.fit_resample(X_train, y_train)


pipe = Pipeline(
    [("minmax", MinMaxScaler()), ("clf", RandomForestClassifier(random_state=42)),]
)

param_grid = {
    "clf__n_estimators": [200, 250, 300, 350],
    "clf__criterion": ["entropy", "gini"],
}

gs = GridSearchCV(pipe, param_grid, verbose=100, cv=10, n_jobs=-2)
gs.fit(X_train, y_train)

gs
````

````python
print(f"Mejor Modelo Grid Search: {gs.best_estimator_}\n")

print(f"Mejor Accuracy Grid Search: {gs.best_score_}\n")

best_random_forest = gs.best_estimator_
y_pred = best_random_forest.predict(X_test)

print("Reporte de Clasificación\n")
print(classification_report(y_test, y_pred))
````

Y use `dump` de `joblib` para persistir el modelo recién entrenado en el disco.

````python
from joblib import dump, load

dump(gs, "modelo.joblib")
````

> ¿Qué es el modelo Cliente-Servidor?   

Corresponde a una arqitectura de software que tiene como principal característica y tal como su nombre lo dice, la separación de las tareas de los administradores de un servicio web y el cliente. El servidor es el encargado de administrar las solicitudes de los clientes y retornar lo deseado. Un buen servidor debe tener capacidad para procesar varias solicitudes de distintos clientes a la vez.  

> ¿Qué es una URI?

  La URI corresponde a las siglas del inglés *Uniform Resource Indentifier*, corresponde a una referencia hacia un servicio web la cual especifica la ubicación y el mecanismo de recuperación.       
    
    
> ¿Qué es HTTP?¿Qué significa que sea stateless?

Protocolo de transferencia de hipertexto (hypertext transfer protocol en inglés), corresponde al protocolo de comunicación para transferencia de información en forma de hipertexto (HTML/XML)


> ¿En que consisten las operaciones de HTTP?: 
   
- GET: Solicita la recuperación de la información solicitada (no hay modificación en el servidor). Puede ser invocada direcamente en una URL.

- POST: Envía información al servidor a través de una sección llamada *body* o cuerpo, usualmente usado para procesar los datos y guardarlos.

- PUT: Similar al POST, pero es usado para actualizar datos previamente existentes.
        
- DELETE: Elimina un dato.

> ¿Qué indican los códigos de respuesta de HTTP?

Los códigos de respuesta indican el éxito o fracaso de una consulta y de qué tipo fue este. Genéricamente hay 5 familias de estas respuestas: 1XX, 2XX, 3XX, 4XX, 5XX.

- 1XX: Corresponden a códigos de respuesta informativos dados durante el proceso de la petición. Un código 1XX se da cuando aún falta por procesar.

- 2XX: Indica el correcto procesamiento de la petición.

- 3XX: Informan la redirección del sitio. En sitios web es fundamental para el SEO (search engine optimization).

- 4XX: Informan un error de parte de quien realiza la petición, ya sea por error de sintaxis, por falta de pago o porque pide café a una tetera, entre otros.

- 5XX: Informa un error de parte del servidor, se da usualmente en la etapa de desarrollo. El cliente no tiene mucho que hacer ante estos errores.

> ¿Qué es un Hipermedio como HTML? 

  Es el medio mediante el cual se representa la información y es transferida a través de http. 


> ¿Qué es una API?

        Siglas de *Application Programming Interface*, que a grandes rasgos no es más que el conjunto de fucniones que provee un servidor al cliente.

> ¿Cómo API REST engloba los términos anteriores?

  La arquitectura REST (Transferencia de estado representacional) engloba a los conceptos anteriores dado que tiene la ventaja de usar HTTP como protocolo de comunicación, luego una API REST es llamable mediante una sintaxis universal a través de la URL con protocolo de transferencia HTTP.

#### Consultas

Podemos hacer consultas, es decir, obtener algunos archivos según algún criterio, a través un objeto `Query` y `search`:

````python
Mediciones = Query()

# Aquí Mediciones.Day representa a todas las llaves Day de todos los documentos.
# luego, el criterio de comparación es el que indica que datos se traeran.
# en la práctica, trae 40 datos, del día 14 al 15.
db.search(Mediciones.Day <= 15)
````

Además, podemos mezclar distintos criterios de forma muy similar como lo hacemos en pandas:

En este caso, la siguiente consulta entrega todas las mediciones con ph entre 6 y 6.5 incluidas.

````python
Mediciones = Query()

db.search((Mediciones.ph >= 6) & (Mediciones.ph <= 6.5))
````

### Update

Podemos actualizar datos que hayan sido seleccionados usando una opración y una `Query` como argumentos de `update`.

Las operaciones disponibles son:

    delete(key): delete a key from the document
    increment(key): increment the value of a key
    decrement(key): decrement the value of a key
    add(key, value): add value to the value of a key (also works for strings)
    subtract(key, value): subtract value from the value of a key
    set(key, value): set key to value


````python
from tinydb.operations import set

Mediciones = Query()

# en este caso, usaremos las mediciones del día 14 y fijaremos su ph a 7
db.update(set("ph", 7), Mediciones.Day == 14)
# noten que esta función retorna los indices de los documentos que fueron actualizados.
````

````python
db.search(Mediciones.Day == 14)
````

### Eliminado


Por último, podemos eliminar de la BBDD con una `Query` y la función `remove`: 

````python
db.remove(Mediciones.Day == 14)
# noten que esta función retorna los indices de los documentos que fueron borrados.
````

````python
db.search(Mediciones.Day == 14)
````

### NOTA: Mutación de la BBDD y Respaldo

Cada vez que ejecutan insert, update o remove modifican las bases de datos. El archivo `db_resplado.json` contiene un respaldo de la base de datos para que puedan volver a utilizar la original.
