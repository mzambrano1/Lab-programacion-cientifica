````python
def agrupar_etnias_padres(df_grades):
  n = df_grades.shape[0]
  grouped = df_grades.groupby(['race/ethnicity','parental level of education'])
  promedio = grouped.mean()
  promedio['porcentaje'] = grouped.count()['names'] * 100 / n
  return promedio
````
